#### Exercise 6.2

This exercise is related to exercise 5.2 of the previous week. Consider the data from CSV file `weight-height.csv`.

1) Read data into a pandas dataframe.

2) Pick the target variable `y` as weight in kilograms, and the feature variable `X` as height in centimeters.

3) Split the data into training and testing sets with 80/20 ratio.

4) Scale the training and testing data using normalization and standardization.

4) Fit a KNN regression model with `k=5` to the training data without scaling, predict on unscaled testing data and compute the $R^2$ value.

6) Repeat step 4 for normalized data.

7) Repeat step 4 for standardize data.

8) Compare the models in terms of their $R^2$ value.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score

#1
df = pd.read_csv("data_banknote_authentication.csv")
print(df)

#2
y = df["class"]
X = df.drop("class", axis=1)
print("X shape:", X.shape)
print("y shape:", y.shape)

#3
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=20)
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)

#4
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

#5
y_pred = svc.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(report)

#6
svc_rbf = SVC(kernel='rbf')
svc_rbf.fit(X_train, y_train)

y_pred_rbf = svc_rbf.predict(X_test)

cm_rbf = confusion_matrix(y_test, y_pred_rbf)
print("Confusion Matrix (RBF Kernel):")
print(cm_rbf)

report_rbf = classification_report(y_test, y_pred_rbf)
print("\nClassification Report (RBF Kernel):")
print(report_rbf)

accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
print("Accuracy (RBF Kernel):", accuracy_rbf)

#7
"""
-the linear SVM assume that classes can be seperate by straight hyperline
-RBF SVM has non linear boundries which map the data into higher dimensional space
-For the Banknote dataset, a linear kernel is very well because it is linearly seperated.
-in case of confusion matrices,fewer off-diagonal entries indicate fewer misclassifications.
-Both SVM models achieved high classification performance on the banknote dataset.
-The RBF kernel has non-linear structure in the data and improvement was small.
-linear SVM remains an attractive choice due to its simplicity and lower computational cost."""
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

#1
df = pd.read_csv("weight-height.csv")
print(df)

#2
X = df[['Height']]
y = df['Weight']
print("X shape:", X.shape)
print("y shape:", y.shape)

#3
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=20)
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)

#4
normalizer = MinMaxScaler()
X_train_norm = normalizer.fit_transform(X_train)
X_test_norm = normalizer.transform(X_test)
print("Normalized training data:")
print(X_train_norm[:5])

standardizer = StandardScaler()
X_train_std = standardizer.fit_transform(X_train)
X_test_std = standardizer.transform(X_test)
print("Standardized training data:")
print(X_train_std[:5])

#5
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R^2 value (unscaled data):", r2)

#6
knn_norm = KNeighborsRegressor(n_neighbors=5)
knn_norm.fit(X_train_norm, y_train)
y_pred_norm = knn_norm.predict(X_test_norm)
r2_norm = r2_score(y_test, y_pred_norm)
print("R^2 value (normalized data):", r2_norm)

#7
knn_std = KNeighborsRegressor(n_neighbors=5)
knn_std.fit(X_train_std, y_train)
y_pred_std = knn_std.predict(X_test_std)
r2_std = r2_score(y_test, y_pred_std)
print("R^2 value (standardized data):", r2_std)

#8
"""
-UNSCALED KNN MODEL:there is only one features "Height" , scaling has little effect because 
 distacne are based on single variable.
-NORMALIZED KNN MODEL:scales height to a fixed range (usually 0-1).
 The distances between samples remain the same, R^2 value is very similar to the unscaled.
-STANDARDIZED KNN MODEL: scales the data based on mean and standard deviation.
 Typically gives performance close to the other two models.
-The three KNN regression models has similar R^2 values ,the model uses only one feature (height).
-The model with the highest R^2 value would be considered the best among the three."""
```
